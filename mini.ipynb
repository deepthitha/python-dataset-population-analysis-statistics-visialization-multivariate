NORMALITY TEST

```python
from scipy.stats import normaltest
columndata= p['Population']
stat, pc = normaltest(columndata)
print("stat=%.10f,p=%.10f" % (stat,pc))
if pc>0.05:
  print("\n normally distributed")
else:
  print("\n not normally distributed")  
```

```python
from scipy import stats
from statsmodels.stats import weightstats as stests
ztest ,pval = stests.ztest(p['Population'], x2=None, value=7.609252e+07)
print(float(pval))
if pval<0.05:
    print("reject null hypothesis")
else:
    print("failed to reject null hypothesis")
```
